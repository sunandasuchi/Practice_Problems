## **617. Merge Two Binary Trees**


```
class TreeNode:
  def __init__(self, val ):
    self.val = val
    self.left = None
    self.right = None
```

```
def merge1(t1,t2):
    
    if t1 == None:
      return t2
    if t2 == None:
      return t1
    t1.val = t1.val + t2.val
    

    t1.left = merge1(t1.left, t2.left)
    
    t1.right = merge1(t1.right, t2.right)
    

    return t1

def inorder(node):
    if (not node):
        return
 
    inorder(node.left) 
 
    print(node.val, end = " ") 
 
    inorder(node.right)

```

```

root1 = TreeNode(100) 
root1.left = TreeNode(2) 
root1.right = TreeNode(3) 
root1.left.left = TreeNode(4) 
root1.left.right = TreeNode(5) 
root1.right.right = TreeNode(6) 
root1.left.left.left = TreeNode(220)
 

root2 = TreeNode(400) 
root2.left = TreeNode(1) 
root2.right = TreeNode(7) 
root2.left.left = TreeNode(3) 
root2.right.left = TreeNode(2) 
root2.right.right = TreeNode(6)
root2.right.right.right = TreeNode(150)


r3 = merge1(root1,root2)

inorder(r3)
```
